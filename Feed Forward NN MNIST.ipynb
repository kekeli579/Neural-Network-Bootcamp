```python
import torch
import torch.nn as nn
import torchvision.datasets as datasets
import torchvision.transforms as transforms
from torch.autograd import Variable
```

```python
input_size = 784 # number of input neurons (28x28 pixels)
hidden_size = 400
out_size = 10
epochs = 10
batch_size = 100
learning_rate = 0.01  #how fast we are learning, when this number is high we will not learn anything
```

```python
train_dataset = datasets.MNIST(root ='./data', train = True, transform = transforms.ToTensor(), download=True)

test_dataset = datasets.MNIST(root ='./data', train = False, transform = transforms.ToTensor(), )
```

```python
#Shuffle for iteration
train_loader = torch.utils.data.DataLoader(dataset= train_dataset, 
                                           batch_size = batch_size, shuffle=True)
test_loader = torch.utils.data.DataLoader(dataset= test_dataset, 
                                           batch_size = batch_size, shuffle=True)
```

```python
#define class (we need to start with initialization function)

class Net(nn.Module):
    def __init__(self, input_size, hidden_size, out_size):
        super(Net, self).__init__()
        self.fc1 = nn.Linear(input_size, hidden_size)    #first layer
        self.relu = nn.ReLU()                           #use activation Relu for other layers
        self.fc2 = nn.Linear(hidden_size, hidden_size)  #2nd layer 
                                            #why we did not create an activation function between the hidden layers?? 
                                            #one attribute (activate function) and it can be use many time)
        self.fc3 = nn.Linear(hidden_size, out_size)     #2nd layer activiation
        
    def forward(self, x):                   #Forward propagate
        out = self.fc1(x)
        out = self.relu(out)
        out = self.fc2(out)
        out = self.relu(out)
        out = self.fc3(out)
        
        return out
```

```python
#create an object of the class, which represents our network
net = Net(input_size, hidden_size, out_size) #class with 3 arguments
CUDA = torch.cuda.is_available()
if CUDA:
    net = net.cuda()
#loss function- cross entropy loss comes with a softmax in pytorch, they are define together
criterion = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(net.parameters(), lr=learning_rate) #Adam optimizer
```

```python
#visualize the train loader (gray scale image, 1 channel of that tensor, 28 # of row and 28 # of colums)
for i, (images, labels) in enumerate(train_loader):
    print(images.size())
    images = images.view(-1, 784)
    print(images.size())
```

```python
#training in pytorch is different than keras (pytorch does not offer a function that directly train the model in the background)
#we train the model from scrap
correct_train = 0
total_train = 0
for epoch in range(epochs):
    for i, (images, labels) in enumerate(train_loader):     #enumerate give the index of every iteration
        images = Variable(images.view(-1, 28*28))           #wrap images in a variable
        labels = Variable(labels)                           #wrap labels in a variable
        if CUDA:
            images = images.cuda()
            labels = labels.cuda()
            
        optimizer.zero_grad()
        outputs = net(images)
        _, predicted = torch.max(outputs.data, 1)           #index location of each maximum value found (max acrross each row for 1)
                                                            # 0 is max accross each column
        total_train += labels.size(0)
        
        if CUDA:
            correct_train += (predicted.cpu() == labels.cpu()).sum()  #predicted is the digit and the label
        else:
            correct_train += (predicted == labels).sum()              #how many correct classify sample got, .sum is a python function
        loss = criterion(outputs, labels)                         ##ce between the actual and predicted
        loss.backward()                                           #backpropagation
        optimizer.step()                                          #update the weights
        
        if (i+1)% 100 ==0:
            print('Epoch[{}/{}], Iteration[{}/{}], Training Loss: {}, Training Accuracy: {}%'.format
            (epoch+1, epochs, i+1, len(train_dataset)//batch_size, loss.data, (100*correct_train/total_train)))
print("Done Training")
```

```python
#Test the network(no loss and weight calculation , no weight update)
correct = 0
total = 0
for images, labels in test_loader:
    images = Variable(images.view(-1,28*28))          #-1 can be replace by bacth number
    if CUDA:
        images = images.CUDA
    outputs = net(images)
    _, predicted = torch.max(outputs.data, 1)         #increment the total count(100)
    total +=labels.size(0)
    
    #we can also use: correct += predicted.eq(labels).sum()
    if CUDA:
        correct += (predicted.cpu() == labels.cpu()).sum()
    else
        correct += (predicted == labels).sum()
        
print('Final Test Accuracy: %d %%' % (100 * correct / total))
```
